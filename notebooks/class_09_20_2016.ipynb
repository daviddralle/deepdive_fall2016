# Electric field due to two point charges separated by 2 unit distance

```python
#import libraries

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import cm
import numpy as np
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15, 15)
```

```python
# We want to manipulate charge values to watch the effect on the electric field. 

from ipywidgets import widgets
from IPython.html.widgets import *

# define the domain, initialize grid

xmin = -2
xmax = 2
nx = 50
ymin = -2
ymax = 2.0
ny = 50
x = np.linspace(xmin,xmax,nx)
y = np.linspace(ymin,ymax,ny)
X, Y = np.meshgrid(x,y)

# define the function that plots the electric field
def pltfield(q1, q2, q2negative): 
    e1 = q1/((X+1)**2+Y**2)*[X+1, Y]/np.sqrt((X+1)**2+Y**2)
    e2 = q2/((1-X)**2+Y**2)*[X-1, Y]/np.sqrt((X-1)**2+Y**2)
    if q2negative:
        e2 = -e2
    field = e1 + e2 
    plot1 = plt.figure()
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.streamplot(X, Y, field[0], field[1], arrowstyle='->',arrowsize=1.5,cmap=cm.cool)
    plt.scatter([-1,1],[0,0])
    ax.text(-1,-.2,'Q1',fontsize=20)
    ax.text(1,-.2,'Q2',fontsize=20)
    
# use interactive widget to watch how e-field changes as a function of charge magnitude and sign
interact(pltfield, q1=FloatSlider(min=.1, max=1, step=.1,continuous_update=False, value=.1),
         q2=FloatSlider(min=.1, max=1, step=.1,continuous_update=False, value=.6), 
         q2negative=Checkbox(value=True,description='Make Q2 negative')
        )

```
