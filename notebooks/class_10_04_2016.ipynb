# Plot electric potential due to dipole

```python
#import libraries

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import cm
import numpy as np
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10, 8)
```

```python
x = np.linspace(-5,5,500)
q = 1
eps = 1
a = 1
v = 1/(4*np.pi*eps)*(q/(np.abs(x-a)) - q/np.abs(a+x))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(x,v)
plt.scatter([-a, a], [0,0])
ax.text(-a,.2,'-Q',fontsize=20)
ax.text(a,-1,'+Q',fontsize=20)
```
